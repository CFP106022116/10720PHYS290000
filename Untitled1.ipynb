```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

im0 = mpimg.imread('0.jpg')
im1 = mpimg.imread('1.jpg')
im2 = mpimg.imread('2.jpg')
im3 = mpimg.imread('3.jpg')
im4 = mpimg.imread('4.jpg')



#print(b_im2.shape)
#print(b_im3.shape)
#print(b_im4.shape)

#plt.imshow(im1)
b_im2 = im2[:,:,2]
b_im3 = im3[:,:,2]
b_im4 = im4[:,:,2]
b_im = np.stack(( b_im2, b_im3, b_im4),axis=0)
b_avg = np.mean(b_im, axis = 0)
plt.imshow(b_avg)

```
